### Read all names

```python
colnames=['name', 'gender', 'count'] 
years = range(1880, 2017, 1)
data = []

for y in years:
  fn = f'yob{y}.txt'
  df = pd.read_csv(fn, names=colnames,header = None)
  df['year'] = y
  data.append(df)

df = pd.concat(data,ignore_index=True)
 
df.tail(10)
```

### Plot a time series


```python
df[df['name'].str.contains("Steffen")]
```

```python
import matplotlib.pyplot as plt

# filter for my name
df_my = df[df['name'].str.contains("Steffen")]

# reindex on year to make time series
df_my2 = df_my.reset_index()
df_my2.set_index('year', inplace=True)

# plot
plt.title('# of Steffen born')
plt.xlabel('year')
plt.ylabel('births')
df_my2['count'].plot(linewidth=3.0 , color='red')

plt.savefig('US_birth_steffen.png', dpi=600)
```

### Normalize

```python
df['totalcount'] = df.groupby('year')['count'].transform('sum')
```

```python
df['frequency'] = df['count'] / df['totalcount']
df.head(10)
```

```python
df_freq = df[(df['name'].isin(['Steffen','Dwayne'] )) & (df['gender'] == 'M')].copy()
df_freq2 = df_freq.set_index([ 'year','name'])
df_freq3 = df_freq2['frequency'].unstack(1)
df_freq3.plot()
```

#### Name diversity

##### one person

```python
#df['unique_names'] = df.groupby('year')['count'].transform('count')
df_agg = df.groupby(['year','gender']).sum().copy()
df_agg.drop(['totalcount', 'frequency'], axis=1, inplace=True)
#df_agg.head(5)
new_df = pd.merge(left=df,right=df_agg,  on =('year','gender'))
#new_df.head(5)
new_df['1_in_x_person'] = new_df['count_y'] / new_df['count_x'] 
#new_df.head(5)
```

```python
df_div = new_df[(new_df['name'] == "Dwayne")& (df['gender'] == 'M')]
# reindex on year to make time series
df_div2 = df_div.reset_index()
df_div2.set_index('year', inplace=True)
df_div2.drop(['index', 'gender','totalcount','frequency','count_y','count_x'], axis=1, inplace=True)
df_div2.head(5)
```

```python
df_div2.plot()
plt.gca().invert_yaxis()
```

##### average number of people with same name per year

```python
df_div4 = new_df.groupby(['year','gender']).mean().copy()
df_div4.drop(['totalcount','frequency','count_x','count_y'], axis=1, inplace=True)
#df_div4 = df_div4.reset_index()
#df_div4.set_index('year', inplace=True)
df_div5 = df_div4.unstack()
df_div5.plot()
plt.gca().invert_yaxis()
```
